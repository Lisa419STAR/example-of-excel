```python
import pandas as pd

# This command loads the xlsx file into a Pandas dataframe
df = pd.read_excel('time_temp.xlsx')
print(df)
```

```python
import matplotlib.pyplot as plt

# We can plot by pulling the data from the datframe and passing it to plt.plot (for example)
plt.plot(df['Time'], df['Temperature'])
```

```python
# Pandas datframes also have an interface to matplotlib built in
# There's some different syntax to learn here, but the basics are straighforward
df.plot('Time', 'Temperature')
```
